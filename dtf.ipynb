```python
# 合併表格並指定列名稱
df = pd.concat([df1, df2], ignore_index=True, axis=0)


# 顯示合併後的前幾行和總行數
print("合併後的數據：")
print(df.head())  # 顯示前五行
print(f"總行數: {len(df)}")
```

```python
# 確保欄位名稱無空白字符
df.columns = df.columns.str.strip()

# 將時間欄位轉為 datetime 格式，並指定正確的格式
df['Time'] = pd.to_datetime(df['Time'], format='%Y/%m/%d/%H:%M:%S', errors='coerce')

# 依據時間升序排序，並移除重複的 Lot_Num，只保留第一次出現的記錄
df_first_sample = df.sort_values(by='Time').drop_duplicates(subset='Lot_Num', keep='first')

# 查看結果
df_first_sample
```

```python
import pandas as pd

# 假設有一個 DataFrame
data = {
    'Lot_Num': ['QW123', 'QW124', 'QW125'],
    'Start_Time': ['2024/10/10/11:43:49', '2024/10/10/12:44:00', '2024/10/10/15:44:10'],
    'End_Time': ['2024/10/12/14:43:49', '2024/10/11/13:44:00', '2024/10/12/16:44:10']
}

df = pd.DataFrame(data)

# 將時間欄位轉換為 datetime 格式
df['Start_Time'] = pd.to_datetime(df['Start_Time'], format='%Y/%m/%d/%H:%M:%S', errors='coerce')
df['End_Time'] = pd.to_datetime(df['End_Time'], format='%Y/%m/%d/%H:%M:%S', errors='coerce')

# 計算時間差
df['Time_Difference'] = df['End_Time'] - df['Start_Time']

# 將時間差轉換為天、時、分的格式並合併成一個欄位
df['Time_Difference_Formatted'] = df['Time_Difference'].apply(lambda x: f"{x.days}天 {x.components.hours}時 {x.components.minutes}分")

# 查看結果
print(df[['Lot_Num', 'Start_Time', 'End_Time', 'Time_Difference_Formatted']])
```
